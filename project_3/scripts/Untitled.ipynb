```python
en_ex["Entry Date"] = en_ex["Entry Date"].fillna("0000")
en_nan = en_ex[(en_ex["Entry Date"] == "0000")].index.tolist()
en_ex = en_ex.drop(en_nan)
en_ex["Exit Date"] = en_ex["Exit Date"].fillna("0000")
ex_nan = en_ex[(en_ex["Exit Date"] == "0000")].index.tolist()
en_ex = en_ex.drop(ex_nan)
```

```python
en_ex["Entry Date"] = pd.to_datetime(en_ex["Entry Date"])
en_ex["Exit Date"] = pd.to_datetime(en_ex["Exit Date"])
en_ex["stay time"] = en_ex["Exit Date"] - en_ex["Entry Date"]
```

```python
date_period = []
stay_time = []
for i in list(range(2015,2019)):
    for j in list(range(1,13)):
        period = str(i)+"-"+str(j)
        start_date = str(i)+"-"+str(j)+"-"+"1"
        if j == 12:
            end_date = str(i+1)+"-"+"1"+"-"+"1"
        else:
            end_date = str(i)+"-"+str(j+1)+"-"+"1"
        date_period.append(period)
        stay_time.append(en_ex.loc[(en_ex["Entry Date"] > start_date) & (en_ex["Entry Date"] < end_date)]["stay time"].mean())
```

```python
stay_time = pd.Series(stay_time).apply(lambda x:int(str(x).split()[0]))
```

```python
#compute how long each client stay each time, add it as a new column "stay time"
def parse_ymd(s):
    mon_s, day_s, year_s = s.split("/")
    return datetime(int(year_s), int(mon_s), int(day_s))
stay_time = []
entry_date = []
exit_date = []
for i in range(len(en_ex)):
    if pd.isnull(en_ex["Entry Date"][i]) == False and pd.isnull(en_ex["Exit Date"][i]) == False:
        entry_date.append(parse_ymd(en_ex["Entry Date"][i]))
        exit_date.append(parse_ymd(en_ex["Exit Date"][i]))
        if en_ex["Entry Date"][i] != en_ex["Exit Date"][i]:
            stay_time.append(int(str(parse_ymd(en_ex["Exit Date"][i]) - parse_ymd(en_ex["Entry Date"][i])).split()[0]))
        else:
            stay_time.append(0)
    else:
        entry_date.append(0)
        exit_date.append(0)
        stay_time.append(0)
en_ex["entry date"] = entry_date
en_ex["exit date"] = exit_date
en_ex["stay time"] = stay_time
df = en_ex[["Client ID","entry date","stay time"]]
df = df.drop(df[(df["entry date"]==0)].index.tolist())
```
